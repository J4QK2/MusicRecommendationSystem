```python
def isNumeric(df):
    result = {}
    
    for column, data in df.items():
        try:
            result[column] = abs(data.values)
        except:
            result[column] = "not numeric"
    
    return result
```

```python
final_artists_df = df_by_artist.head(7000).copy()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(final_artists_df['artists'].head(7000))

tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

print(tfidf_df)
print("The number of non zero names: ", np.count_nonzero(tfidf_df.values))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_matrix = cosine_similarity(tfidf_df)
```

```python
print(similarity_matrix)
```

```python
similar = similarity_matrix[4]
top_indices = np.argsort(similar)[::-1][1:6]
```

```python
final_artists_df.iloc[top_indices]
```

```python
def isSimilar(artist, song):
    # объединяем как в датасете
    text = artist + " " + song
    
    # превращаем в TF-IDF вектор
    vector = vectorizer.transform([text])
    
    # считаем сходство с базой
    similarities = cosine_similarity(vector, tfidf_matrix)
    
    # находим самые похожие
    top_indices = np.argsort(similarities[0])[::-1][:5]
    
    return final_artists_df.iloc[top_indices]
```

```python
isSimilar("Eazy-E", "Diva")
```
