```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer


vectorizer = TfidfVectorizer()

countVectorizer = CountVectorizer(max_features=20000)

tfidf_matrix = vectorizer.fit_transform(final_artists_df['features']).toarray()

count_matrix = countVectorizer.fit_transform(final_artists_df['features']).toarray()

#tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

#print(tfidf_df)
#print("The number of non zero names: ", np.count_nonzero(tfidf_df.values))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_matrix = cosine_similarity(count_matrix)
```

```python
def isSimilar(music):
    #проверяет есть ли похожое название музыки, в противном случае вызовет пустую строку
    music_row = final_artists_df[final_artists_df['name'].str.contains(music, case=False, na=False)]
    
    if music_row.empty:
        print("Song not found")
        return
    
    #размер необходимой строки и схожесть между значениями в этой строке
    music_index = music_row.index[0]
    distances = similarity_matrix[music_index]

    print(music_index)

    #сортировка списка с рекомендацией, начиная с самого похожего, берем лишь первые пять песен, и выводим название x[1]
    music_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:6]
    
    for i in music_list:
        print(final_artists_df.iloc[i[0]]['name'])
    
```

```python
isSimilar('gay')
```
